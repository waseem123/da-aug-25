```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('emp-dataset.csv')
```

```python
df
```

```python
# Creating a Dashboard Layout with Title
fig,axs=plt.subplots(2,2,figsize=(12,8))
fig.suptitle('Employee Sales Dashboard')

# Titles for every graph
axs[0,0].set_title('Sales vs Target')
axs[0,1].set_title('Projects Handled')
axs[1,0].set_title('Sales Contribution')
axs[1,1].set_title('Summary')

# Bar graph for Sales Vs Targets
axs[0,0].bar(df['employee'],df['sales'], label='Sales',alpha=0.7,color='Green')
axs[0,0].bar(df['employee'],df['targets'],label='Targets',alpha=0.7,color='Blue')
axs[0,0].legend()

# Scatter Plot for Handled Projects
axs[0,1].plot(df['employee'],df['projects'],marker='o',label='Projects')
axs[0,1].legend()

# Pie chart for Sales Contribution
axs[1,0].pie(df['sales'],labels=df['employee'],autopct='%1.1f%%',startangle=90)
axs[1,0].legend()

#Table for summary of all graphs
summary = df[['sales','targets','projects','satisfaction']].describe().round(2)
axs[1,1].axis('tight')
axs[1,1].axis('off')
axs[1,1].table(rowLabels=summary.index,colLabels=summary.columns,cellText=summary.values,loc='center')

plt.tight_layout(rect=[0,0,1,0.95])
plt.savefig('Employee_Sales_dashboard.png')
plt.show()
```
